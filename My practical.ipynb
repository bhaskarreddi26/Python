```
import pyspark.sql.types as typ

labels =[
  ('satisfaction_level',typ.DecimalType()),
  ('last_evaluation',typ.DecimalType()),
  ('number_project',typ.IntegerType()),
  ('average_montly_hours',typ.IntegerType()),
  ('time_spend_company',typ.IntegerType()),
  ('Work_accident',typ.IntegerType()),
  ('left',typ.IntegerType()),
  ('promotion_last_5years',typ.IntegerType()),
  ('sales',typ.StringType()),
  ('salary',typ.StringType())
]
schema = typ.StructType([
    typ.StructField(e[0], e[1], False) for e in labels
])

hrpeople = spark.read.csv('/FileStore/tables/HR_comma_sep.csv', 
                        header=True, 
                        schema=schema)
```

```
hrpeople.show(2)
```

```
from pyspark.sql import functions as F
hrpeople.select(hrpeople.number_project, F.when(hrpeople.left==1 , 0).otherwise(1)).show()
```
