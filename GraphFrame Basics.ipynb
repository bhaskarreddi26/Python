Creating Dataframe for vertices & edges.

```
v = sqlContext.createDataFrame([
  ("a", "Alice", 34),
  ("b", "Bob", 36),
  ("c", "Charlie", 30),
  ("d", "David", 29),
  ("e", "Esther", 32),
  ("f", "Fanny", 36),
  ("g", "Gabby", 60)
    
], ["id", "name", "age"])


# Edge DataFrame
e = sqlContext.createDataFrame([
  ("a", "b", "friend"),
  ("b", "c", "follow"),
  ("c", "b", "follow"),
  ("f", "c", "follow"),
  ("e", "f", "follow"),
  ("e", "d", "friend"),
  ("d", "a", "friend"),
  ("a", "e", "friend")
], ["src", "dst", "relationship"])
```

```
display(e)
```

```
from graphframes import *

g = GraphFrame(v, e)
```

```
v2 = g.vertices.filter("age > 30")
e2 = g.edges.filter("relationship = 'friend'")
g2 = GraphFrame(v2,e2)
```

```
display(e2)
```

```
print g2.degrees.collect()
print g2.outDegrees.collect()
print g2.inDegrees.collect()
print g2.triangleCount().show()
```
